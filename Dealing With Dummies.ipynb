```python
import pandas as pd
df= pd.read_csv("C:/Users/LILENDAR/Desktop/PROJECT 1/College1234.csv")
```

```python
d={"1Star":1,"2Star":2,"3Star":3,"4Star":4,"5Star":5}
df["University"]=df["University"].apply(lambda df:d[df])
```

```python
df.head()
```

```python
data3=pd.get_dummies(df, columns=['Research'])
data3.head()
```

```python
data3.columns
```

```python
import statsmodels.formula.api as sm
model1 = sm.ols(formula='Chance~GRE+TOEFL+University+SOP+LOR+CGPA+Research_0+Research_1', data=data3).fit()
model1.summary()
```
