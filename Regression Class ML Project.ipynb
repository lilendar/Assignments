```python
import pandas as pd
import numpy as np
data=pd.read_csv("C:/Users/LILENDAR/Desktop/PROJECT 1/data.csv")
```

```python
data.head()
```

```python
data2= data.loc[:, ~data.columns.str.contains('^Unnamed')]
data2.head()
```

```python
import seaborn as sns
corr = data2.corr()
print(corr)
```

```python
import matplotlib.pyplot as plt
labels = data2['Y1']
features = data2.drop(['Y1'], axis=1)

for x in features:
    plt.plot(labels, features[x], 'ro')  # arguments are passed to np.histogram
    plt.title("Y1 vs " + x)
    plt.xlabel(x)
    plt.ylabel("Y1")
    plt.show()
```

```python
for x in features:
    plt.hist(features[x], bins='auto')  # arguments are passed to np.histogram
    plt.title(x)
    plt.show()
```

```python
from scipy import stats
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(211)
x = data2['X4']
prob = stats.probplot(x, dist=stats.norm, plot=ax1)
ax1.set_xlabel('')
ax1.set_title('Probplot against normal distribution')
#We now use boxcox to transform the data so its closest to normal:
ax2 = fig.add_subplot(212)
data2['X4'], _ = stats.boxcox(x)
prob = stats.probplot(data2['X4'], dist=stats.norm, plot=ax2)
ax2.set_title('Probplot after Box-Cox transformation')
plt.tight_layout()
plt.show()
```

```python
plt.hist(data2['X4'], bins='auto')  # arguments are passed to np.histogram
plt.title("X4 after Box cox transformation")
plt.show()
```

```python
plt.plot(data2['Y1'], data2['X4'], 'ro')  # arguments are passed to np.histogram
plt.title("Scatter plot b/w Y1 and X4")
plt.xlabel("X4")
plt.ylabel("Y1")
plt.show()
```

```python
import statsmodels.formula.api as sm
model1 = sm.ols(formula="Y1~X1+X2+X3+X4+X5+X6+X7+X8", data=data2).fit()
model2 = sm.ols(formula="Y1~X1+X2+X3+X5+X6+X7+X8", data=data2).fit()
model3 = sm.ols(formula="Y1~X1+X2+X3+X5+X7+X8", data=data2).fit()
model4 = sm.ols(formula="Y1~X1+X2+X3+X5+X7", data=data2).fit()

#sales~TV+radio+newspaper
print(model1.summary())
print(model2.summary())
print(model3.summary())
print(model4.summary())
```

```python
print('Parameters: ', model2.params)
print('R2: ', model2.rsquared)
print('Standard errors: ', model2.bse)
```

```python
y_pred = model4.predict()
print(y_pred)
```
