```python
tables = camelot.read_pdf('short_lines.pdf')
camelot.plot(tables[0], kind='grid')
plt.show()
```

```python
#the small lines are not detected
# here line_scale=40 is used

tables = camelot.read_pdf('short_lines.pdf', line_scale=40)
camelot.plot(tables[0], kind='grid')
plt.show()
```

```python
tables[0].df
```

```python
#placement of the text within the boxes.....where weset the gravity option which shifts the word 
#in free float
#we use shift_text=[""]
#   '', l', 'r', 't', 'b'),


tables = camelot.read_pdf('short_lines.pdf', line_scale=40, shift_text=[''])
tables[0].df
```

```python
#Let’s pass shift_text=['r', 'b'] to set the gravity to right-bottom 
#and move the text in that direction

tables = camelot.read_pdf('short_lines.pdf', line_scale=40, shift_text=['r', 'b'])
tables[0].df
```
