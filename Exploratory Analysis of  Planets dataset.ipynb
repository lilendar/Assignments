# BENCHMARK CREATION

```python
import math
import numpy as np

RMS1= math.sqrt(np.mean((train["x"]-train["x_sim"])**2))
RMS2= math.sqrt(np.mean((train["y"]-train["y_sim"])**2))
RMS3= math.sqrt(np.mean((train["z"]-train["z_sim"])**2))
RMS4= math.sqrt(np.mean((train["Vx"]-train["Vx_sim"])**2))
RMS5= math.sqrt(np.mean((train["Vy"]-train["Vy_sim"])**2))
RMS6= math.sqrt(np.mean((train["Vz"]-train["Vz_sim"])**2))


print("ROOT MEAN DEVIATION IN X CO-ORDINATE = ", RMS1)
print("ROOT MEAN DEVIATION IN Y CO-ORDINATE = ", RMS2)
print("ROOT MEAN DEVIATION IN Z CO-ORDINATE = ", RMS3)
print("ROOT MEAN DEVIATION IN Vx CO-ORDINATE = ", RMS4)
print("ROOT MEAN DEVIATION IN Vy CO-ORDINATE = ", RMS5)
print("ROOT MEAN DEVIATION IN Vz CO-ORDINATE = ", RMS6)
```
