```python
import pandas as pd

df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)

for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='â„–':
        df.rename(columns={col:'#'+col[1:]}, inplace=True)

names_ids = df.index.str.split('\s\(') # split the index by '('

df.index = names_ids.str[0] # the [0] element is the country name (new index) 
df['ID'] = names_ids.str[1].str[:3] # the [1] element is the abbreviation or ID (take first 3 characters from that)

df = df.drop('Totals')
df.head()
```

```python
def answer_one():
    only_gold=df[(df['Gold']>0)&(df['Gold.1']>0)]
    only_gold['Difference']=(abs(only_gold['Gold']-only_gold['Gold.1']))/only_gold['Gold.2']
    return only_gold['Difference'].idxmax()
answer_one()
```

```python
def answer():
    df['Points']=df['Gold.2']*3+df['Silver.2']*2+df['Bronze.2']*1
    return df['Points']
answer()
```

```python
def answer_three():
    k=[]
    data1=df[(df['Gold']>1)&(df['Gold.1']>1)]

    for i in range(len(data1["Gold"])):

    
        diff=data1["Gold"][i]-data1["Gold.1"][i]
        summ=data1["Gold"][i]+data1["Gold.1"][i]
        p=abs(diff/summ)
        k.append(p)
        
    country=data1.index[k.index(max(k))]
    return country
```

```python
answer_three()
```

```python
census_df = pd.read_csv('census.csv')
census_df.head(30)
```

```python
def answer_six(): 
    df1=census_df.sort_values(['STNAME','POPESTIMATE2015'], ascending=False).groupby('STNAME').head(3).copy()
    df2=df1.reset_index().groupby('STNAME').sum().sort_values(['POPESTIMATE2015'], ascending=False).head(3).copy()
    return list(df2.index)
answer_six()
```

```python
def answer_seven():
    df=census_df[census_df['SUMLEV']==50]
    df['STDEV']=df[['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013',
                'POPESTIMATE2014','POPESTIMATE2015']].std(axis=1)
    return df.loc[df['STDEV'].idxmax()]['CTYNAME']
answer_seven()
```

```python
def answer_eight():
    df=census_df[census_df['SUMLEV']==50]
    region1=df['REGION']==1
    region2=df['REGION']==2
    country_name=df['CTYNAME'].str.startswith('Washington')
    population=df['POPESTIMATE2015']>df['POPESTIMATE2014']
    df=df[(region1|region2)&country_name&population]
    return df[['STNAME','CTYNAME']]
answer_eight()
```
